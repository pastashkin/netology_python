### 1 Классическая теория вероятностей

### Ответ 1

(a) Для решения задачи, представим как в общем виде можно расставить ладьи так, чтобы они не били друг друга
Первую ладью можно разместить n способами. Вторую (n-1) и так далее до n-ной ладьи. Учитывая, что доска квадратная получаем, что А = n * (n-1) * (n-2) * ... * 2 * 1 = n!

```python
import math

n = 5
A = math.factorial(n)
print(A)
```

(b) Наскольно  понял, в этом случае, для выбора первого места для ладьи у нас будет n-1 мест и так далее. Т.е. если в прошлом решении у нас было на 1 размещение меньше, в этом будет на 2 (горизонталь и прошлой ладьи и главная диагональ).

```python
n = 5
B = math.factorial(n-1)
print(B)
```

### Ответ 3

Пусть вероятность дожить до 60 P(A) = 1/2 и вероятность дожить до 80 P(B) = 1/5. Поскольку мы наблюдаем условную вероятность, то шанс дожить до 80 при условии, что мы дожили до 60 будет P(B|A) = P(AB) / P(A) = 1/5 / 1/2 = 2/5

### Ответ 4

```python
# Вероятность брака
bad_a = 0.1
bad_b = 0.2
bad_c = 0.15
# Шанс выбрать пару обуви конкретной макри
type_a = 40 / 100
type_b = 30 / 100
type_c = 30 / 100

#(a) Воспользуемся формулой полной вероятности
good_pair = type_a * (1-bad_a) + type_b * (1-bad_b) + type_c * (1-bad_c)
print('Шанс выбрать хорошую пару: ' + str(round(good_pair,3)))

#(b) Воспользуемся формулой Байеса
bad_pair = type_a * bad_a + type_b * bad_b + type_c * bad_c
bad_is_b = bad_b * type_b / bad_pair
print('Шанс выбрать вторую пару, при условии, что выбрана плохая: ' + str(round(bad_is_b,3)))
```

### Ответ 2

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
import math
%matplotlib inline

# Распределение Пуассона
x_1= np.arange(0,20,1)
x_2= np.arange(0,20,1)

l_1 = 5
l_2 = 11

y_1 = []
y_2 = []
for i in x_1:
    y_1.append((l_1**i / math.factorial(i)) * (math.e ** (-1 * l_1)))
for i in x_2:
    y_2.append((l_2**i / math.factorial(i)) * (math.e ** (-1 * l_2)))

plt.xlabel('value')
plt.ylabel('probability')

plt.plot(x_1, y_1, label='$\lambda$=5, k=20')
plt.plot(x_2, y_2, label='$\lambda$=11, k=20')
plt.legend(loc='upper right')

plt.show()
```

```python
# Экспоненциальное
x_1 = np.linspace(0, 10, 100)
y_1 = sts.expon.pdf(x_1)
x_2 = np.linspace(0, 10, 10)
y_2 = sts.expon.pdf(x_2)
plt.plot(x_1, y_1)
plt.plot(x_2+0.5, y_2+0.1)

plt.xlabel('value')
plt.ylabel('probability')

plt.show()
```

```python
# Гамма-распределение
a_1 = 1
a_2 = 2
x_1 = np.linspace(0.01, 4, 1000)
x_2 = np.linspace(0.01, 4, 1000)
plt.plot(x_1, sts.gamma.pdf(x_1, a_1))
plt.plot(x_2, sts.gamma.pdf(x_2, a_2))
plt.xlabel('value')
plt.ylabel('probability')
plt.show()
```
